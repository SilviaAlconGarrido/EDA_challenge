# Primero leemos los datos

```python
df_city = pd.read_csv('./data/diamonds_city.csv')
print(df_city.info(5))
```

```python
df_clarity = pd.read_csv('./data/diamonds_clarity.csv')
print(df_clarity.info(5))
```

```python
df_color = pd.read_csv('./data/diamonds_color.csv')
print(df_color.info(5))
```

```python
df_cut = pd.read_csv('./data/diamonds_cut.csv')
print(df_cut.info(5))
```

```python
df_dimensions = pd.read_csv('./data/diamonds_dimensions.csv')
print(df_dimensions.info(5))
```

```python
df_properties = pd.read_csv('./data/diamonds_properties.csv')
print(df_properties.info(5)) 
```

```python
df_transactional = pd.read_csv('./data/diamonds_transactional.csv')
print(df_transactional.info(5))
```

```python
mergedf1 = df_clarity.merge(df_properties)
mergedf1
```

```python
mergedf2 = df_dimensions.merge(df_transactional)
mergedf2
```

```python
mergedf3 = df_city.merge(df_transactional)
mergedf3
```

```python
mergedf4 = df_cut.merge(df_properties)
mergedf4
```

```python
mergedf5 = df_color.merge(df_properties)
mergedf5
```

```python
df_final = mergedf2.merge(mergedf1).merge(mergedf3).merge(mergedf4).merge(mergedf5).merge(df_dimensions)
df_final
```

```python
df_final2 = df_final[['index_id', 'price', 'carat', 'city', 'cut', 'color', 'depth', 'table', 'x', 'y', 'z', 'clarity']]
df_final2
```
